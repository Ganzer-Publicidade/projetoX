```python
# Importar m√≥dulo
from src.audio_generator import AudioGenerator

# Criar gerador
generator = AudioGenerator(api_key=ELEVENLABS_API_KEY)

print("‚úÖ Gerador de √Åudio inicializado!")
```

```python
# Texto de teste
texto_teste = """
H√° muito tempo atr√°s, na terra de Israel, vivia um jovem pastor chamado Davi.
Ele era corajoso, tinha f√© em Deus e cuidava de suas ovelhas com dedica√ß√£o.
Um dia, um gigante filisteu chamado Golias desafiou o ex√©rcito de Israel.
Ningu√©m tinha coragem de enfrent√°-lo, at√© que Davi se apresentou.
Com apenas uma funda e cinco pedras, Davi confiou em Deus e derrotou o gigante.
Esta √© uma hist√≥ria sobre f√©, coragem e a vit√≥ria do bem sobre o mal.
""".strip()

print("üìù Texto para narra√ß√£o:")
print(texto_teste)
```

```python
# Gerar √°udio
print("\nüéôÔ∏è Gerando narra√ß√£o...")

audio_path = generator.gerar_narracao(
    texto=texto_teste,
    idioma="pt-br",
    voice_id=None,  # Usa voz padr√£o
    output_path="/content/narracao_teste.mp3"
)

print("\n" + "=" * 60)
print("‚úÖ √ÅUDIO GERADO COM SUCESSO!")
print("=" * 60)
print(f"üìÅ Salvo em: {audio_path}")
```

```python
# Reproduzir √°udio
from IPython.display import Audio

print("\nüîä REPRODUZIR √ÅUDIO:\n")
display(Audio(audio_path))
```

```python
# Informa√ß√µes do √°udio
from pydub import AudioSegment

audio = AudioSegment.from_mp3(audio_path)
duracao_segundos = len(audio) / 1000.0
tamanho_kb = os.path.getsize(audio_path) / 1024

print("\nüìä INFORMA√á√ïES DO √ÅUDIO:")
print(f"   ‚è±Ô∏è Dura√ß√£o: {duracao_segundos:.2f}s")
print(f"   üì¶ Tamanho: {tamanho_kb:.2f} KB")
print(f"   üéµ Sample rate: {audio.frame_rate} Hz")
print(f"   üîä Canais: {audio.channels}")
```

```python
# Testar com roteiro completo (exemplo)
roteiro_exemplo = {
    'titulo': 'A Hist√≥ria de Davi e Golias',
    'cenas': [
        {
            'numero': 1,
            'tipo': 'narracao',
            'texto': 'Era uma vez, na antiga Israel, um jovem pastor chamado Davi.',
        },
        {
            'numero': 2,
            'tipo': 'narracao',
            'texto': 'Ele cuidava das ovelhas de seu pai com muito carinho e dedica√ß√£o.',
        },
        {
            'numero': 3,
            'tipo': 'narracao',
            'texto': 'Um dia, um gigante filisteu chamado Golias desafiou Israel.',
        }
    ]
}

print("\nüé¨ Gerando √°udios para roteiro completo...")

audios = generator.gerar_audio_cenas(
    roteiro=roteiro_exemplo,
    idioma="pt-br"
)

print(f"\n‚úÖ {len(audios)} √°udios gerados!")
```

```python
# Salvar cat√°logo
output_path = "/content/audios_catalogo.json"
generator.salvar_catalogo_audios(audios, output_path)

print(f"\n‚úÖ Cat√°logo salvo em: {output_path}")
```

```python
# Estat√≠sticas
total_caracteres = len(texto_teste)
custo_estimado = (total_caracteres / 1000) * 0.30  # $0.30 por 1K chars

print("\nüìä ESTAT√çSTICAS:")
print(f"   üìù Total de caracteres: {total_caracteres}")
print(f"   üí∞ Custo estimado: ${custo_estimado:.3f}")
print(f"   ‚è±Ô∏è Tempo de √°udio: {duracao_segundos:.2f}s")
```
